```python
data_path = '/content/drive/MyDrive/Internship/dummy.csv'
```

```python
dataset = pd.read_csv(data_path)
```

```python
dataset.head(15)
```

```python
dataset.info()
```

```python
dataset['Bulan'] = pd.to_datetime(dataset['Bulan'])
```

```python
dataset.info()
```

```python
dataset.head(15)
```

```python
dataset['BD_percent'] = dataset['Menit'] / dataset['Load_time']
```

```python
dataset.head(15)
```

```python
DGM1 = dataset[dataset['Mesin']=='DGM 1'].reset_index(drop=True)
DGM2 = dataset[dataset['Mesin']=='DGM 2'].reset_index(drop=True)
DGM2K = dataset[dataset['Mesin']=='DGM 2 (KORAN)'].reset_index(drop=True)
DGM3 = dataset[dataset['Mesin']=='DGM 3'].reset_index(drop=True)
GOSS546 = dataset[dataset['Mesin']=='GOSS 546'].reset_index(drop=True)
SOLNA578 = dataset[dataset['Mesin']=='SOLNA 578'].reset_index(drop=True)
SOLNA301 = dataset[dataset['Mesin']=='SOLNA 301'].reset_index(drop=True)
SOLNA630L1 = dataset[dataset['Mesin']=='SOLNA 630 L1'].reset_index(drop=True)
SOLNA630L2 = dataset[dataset['Mesin']=='SOLNA 630 L2'].reset_index(drop=True)
SOLNA630L3 = dataset[dataset['Mesin']=='SOLNA 630 L3'].reset_index(drop=True)
```

```python
# DGM 1
sns.lineplot(x='Bulan', y='Target', linestyle='dashed', color='r', data=DGM1, marker='.', markerfacecolor='black', markersize=10)
sns.lineplot(x='Bulan', y='BD_percent', data=DGM1, marker='.', markerfacecolor='black', markersize=10)
```
